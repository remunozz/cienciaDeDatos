```python
#código que permite calcular el total de camas críticas para cada uno de los servicios de salud

import pandas as pd

# Leer dos columnas específicas de una hoja de un archivo de Excel y el titulo de las columnas necesarias
df = pd.read_excel('datos.xls', sheet_name='Hospitales', usecols=['Nombre Servicio de Salud','Total Camas Críticas'])

suma = df.groupby('Nombre Servicio de Salud').sum()  
# Imprimir el total de camas críticas por servicio
print(suma)
print()
minimo = df.groupby('Nombre Servicio de Salud').min()
# Imprimir el minimo de camas críticas por servicio
print(minimo)
print()
# Imprimir servicios de salud son los tres servicios que tienen más camas críticas 
tresServiciosConMasCamasCriticas = suma['Total Camas Críticas'].nlargest(3)
print(tresServiciosConMasCamasCriticas)
```

```python
# #código que permite calcular el total de camas críticas para cada uno de los servicios de salud
import pandas as pd
# Leer dos columnas específicas de una hoja de un archivo de Excel
df = pd.read_excel('datos.xls', sheet_name='Hospitales')
suma = df[['Nombre Servicio de Salud','Total Camas Críticas']].groupby('Nombre Servicio de Salud').sum()  
# Imprimir el resultado
print(suma)
```

```python
# ¿Cuál cifra corresponde al promedio diario de pacientes que han llegado a la red en los últimos 10 días (desde el día 415 al día 424)? 
import pandas as pd
df = pd.read_excel('datos.xls', sheet_name='Información Histórica')
df_rango = df.loc[(df['Día de llegada'] >= 415)]
conteo = df_rango['Día de llegada'].value_counts()
print(conteo)
print()
print(conteo.mean())
```

```python
# ¿Cuál es el promedio del costo diario de resolución en clínica privada de los 3 GRDs más costos?
import pandas as pd
df= pd.read_excel('datos.xls', sheet_name='GRD')
sorted_df = df.sort_values('Costo resolución en Clínica Privada [MM$]', ascending=False)
top_three = sorted_df.iloc[:3]['Costo resolución en Clínica Privada [MM$]']
promedio = top_three.mean()
print(promedio)
```

```python
# ¿Qué cifra corresponde a la diferencia entre el menor y mayor costo, de operación anual de una cama para cada servicio de salud?
import pandas as pd
df= pd.read_excel('datos.xls', sheet_name='Hospitales')
max = df['Costo de operación anual de una cama [MM$]'].max() 
min = df['Costo de operación anual de una cama [MM$]'].min() 
dif = max - min
print(dif)
```

```python
# Este código permite determinar el valor del costo para instalar 3 camas nuevas y de las menos costosas- en cada servicio de salud
import pandas as pd
df= pd.read_excel('datos.xls', sheet_name='Hospitales')
costo_camas = df[["Nombre Servicio de Salud", "Costo de instalar una cama [MM$]"]].groupby("Nombre Servicio de Salud").min()
print(costo_camas)
print()
costo_camas["Costo 3 camas nuevas"] = costo_camas["Costo de instalar una cama [MM$]"]*3
print(costo_camas)
```
