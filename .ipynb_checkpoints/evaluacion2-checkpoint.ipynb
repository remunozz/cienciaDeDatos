```python
# Pregunta 1
# ¿Cuál opción corresponde al servicio de salud al que llegaron más pacientes el día 1? 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

Info_sheet=pd.read_excel('Datos.xls', sheet_name='Información Histórica', header=0, index_col=None)
Info_sheet['Número de pacientes']=Info_sheet['Día de llegada']

item1 = Info_sheet.loc[Info_sheet['Número de pacientes'] == 1]
numeroDePacientesElDia1PorHospital = item1[['Hospital al que llega','Número de pacientes']].groupby('Hospital al que llega').sum()
print(numeroDePacientesElDia1PorHospital)
print()
print(numeroDePacientesElDia1PorHospital.sort_values(by=['Número de pacientes'], ascending = False))
print()
index_max = numeroDePacientesElDia1PorHospital['Número de pacientes'].idxmax()
result = numeroDePacientesElDia1PorHospital.loc[index_max]
print('Hospital al que llegaron más pacientes el día 1:')
name = result.name
print(name,'con', result[0], 'pacientes.')
print(f'{name} con {result[0]} pacientes.')

Hosp_sheet=pd.read_excel('Datos.xls', sheet_name='Hospitales', header=0, index_col=None)
itmemName=Hosp_sheet.loc[Hosp_sheet['Nombre Establecimiento']==name]
itmemName[['Nombre Servicio de Salud','Nombre Establecimiento']]
```
